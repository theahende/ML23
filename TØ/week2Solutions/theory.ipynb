# 3. The Perceptron
### Question 1: Running the Perceptron Learning Algorithm
Assume we are given a training data set with 3 features, of which the first is hardcoded to 1. The data consists of the four examples $((1,2,2), 1), ((1, 2, 3), 1), ((1, 4, 2), -1), ((1, 4, 0), -1)$. What hypothesis $w = (w_0,w_1,w_2)$ does it return when initialized with $w = (0,0,0)$ and where we always pick the first misclassified point when updating?

### Solution Math
Initially, we have $w = (0,0,0)$ and thus all examples are misclassified. We therefore pick the first and update $w$ to $(1,2,2)$. 

Next, we see that $w=(1,2,2)$ predicts 1 on all 4 examples. Hence the first misclassified example is $((1, 4, 2),-1)$. We therefore update $w$ to $w=(0,-2,0)$.

We then see that $w=(0,-2,0)$ mispredicts the first example $((1,2,2),1)$. We then update $w$ to $(1,0,2)$.

The new $w$ predicts $1$ on all examples. The first misclassified is $((1, 4, 2) ,-1)$. The new $w$ is thus $w=(0,-4,0)$.

The new $w$ mispredicts the first example. We then update $w$ to $w=(1,-2,2)$.

The new $w$ predicts $1$ on the first and second point. It predicts $-1$ on $((1, 4, 2), -1)$ and also on $(1, 4, 0)$. Hence the algorithm terminates.

### End Solution Math

Show that $h_\mathrm{med} = \mathrm{median}(y_1,\dots,y_n)$, the median of the $y$ values is the minimizer.

<b>Question 3: </b>What happens to the solutions $h_\mathrm{mean}, h_\mathrm{med}$ if we

$\varepsilon \rightarrow \infty$.



Which method is more stable for outliers (data that looks nothing like the remaining data)?

### solution math
Least Squares:
$$
f(h) = \sum_{i=1}^n (h-y_i)^2, \quad f'(h) = \sum_{i=1}^n 2(h-y_i)
$$
set to zero
$$
\sum_{i=1}^n 2(h-y_i) = 0 \Rightarrow h = \frac{1}{n} \sum_{i=i}^n y_i
$$

Least Absolute Value:

You can do the same method except you need the gradient of abs which requires subgradients to be formal. For now say it is 0 at 0, -1 for negative x, and +1 for positive x.

Purely intuitive algorithhmic argument: Argue what happens with the cost as we move h from  negative infinity to positive infinity. 
The elements $y$ you move closer to reduce their cost in the sum linearly, i.e. if you move $h$ by $\delta$ than all of these decrease by $\delta$ (unless you pass one of them).
Similar for the points $y$ you move away from. If you are moving closer to more elements than you are moving away from you are decreasing cost, and you are increasing cost if you move away from more points than you move closer to.
Alas you must find a point that has an equal amount of elements that are smaller and larger...... The Median!!!

$$
f(h) = \sum_{i=1}^n |h-y_i|, \quad f'(h) = \sum_{h > y_i} 1 + \sum_{h < y_i} -1
$$
let $a = |\{i : h > y_i\}|, b =|\{i : h < y_i\}|$
For the gradient to be zero it must be the case that $a=b$ which indeed is a local minimum.
Thus you must find $h$ such that the number of $y$ larger than $h$ is equal to the number of $y$ less than $h$. The median.

# 5. Decision Tree Cost with Entropy
In this exercise we examine the entropy-based approach to constructing decision stumps. Recall that, for any leaf $\ell$, the entropy in that leaf is $H(\ell):=-\sum_{i=0}^{k-1} p_i \lg_2(p_i)$. Here $p_i$ denotes the fraction of training examples in that leaf having the label $i$. For binary classification, we thus have $k=2$.

The entropy of the entire tree $T$ is $H(T):=\sum_{\ell} (n_\ell/n) H(\ell)$, where $\ell$ sums over all leaves and $n_\ell$ is the number of training examples in leaf $\ell$.

# 6: Implementing Regression Stumps
In this exercise your task is to implement Regression Trees that consist of one internal node (the root) and two leafs.
Such trees are known as Regression Stumps.
For the loss/cost function we consider least squares loss $(h(x) - y)^2$ 

This means that the learning algorithm has to find the best possible feature to split the training data using a single feature value pair in regards to Least Squares loss.

We have decided for you to present a Regression Stump by 
- idx: the data/feature vector index to consider in the root node (the one question asked)
- val: the value to compare to for data feature idx in the root node
- left: the value to return for a data point if it ends up in left leaf (x[idx] < val) (only question type we consider in a node)
- right: the value to return for a data point if it ends up in the right leaf (x[idx] >= val)

The approach we follow is as follows. Assume the input data has n data points each a vector of $d$ real numbers.

**Basic Algorithm:**

For each data feature f:
-   Compute for all possible values $v$ for feature $f$ in the training data, the least squares cost of the stump achieved by using feature f and value $v$ in the root using the optimal value in the two leafs. This gives a list of of costs, one for each split ($f, v: \textrm{cost}$). 
- Pick the split $f, v$ with minimal cost and create the corresponding tree by setting idx, val, left, right
 
Your task is to give a full implementation of this algorithm and specify the running time.

# 8. BONUS exercise if time: Classification Stumps in O(n d lg n ) time
In this exercise your job is to describe an algorithm that given a data set of labelled data (two classses only), constructs the binary classification tree (one internal node and two leafs) that minimize the 0-1 Loss over the training data. Such small classification trees are called classification stumps.

i.e. given data 
$$ D = \{(x_i, y_i) \mid 1\leq i \leq n, y_i \in \{0, 1\}, x_i \in \mathbb{R}^d\}$$
construct the binary classification tree $T$ that minimize 
$$\frac{1}{n} \sum_{i=1}^n 1_{T(x_i) \neq y_i}$$.
Your algorithm must only use $O(n d \log n)$ time. 

The root node considers only questions like $f_i < 42$ and this may be represented by the feature's index i and the value to compare with (42 here).


**Hint: Consider each feature in turn and sort the data based on that feature. Compute the score for each relevant split in $O(n \lg n)$ time**


### solution math
For a given feature we sort the feature and get the permutation and use that to sort the labels the same way.
Now for each value of the feature in this sorted order we need to compute the cost of using this as root node.
We will scan the values from left to right and maintain the following values for the two leaves $l, r$ (left, right)
- $c_0^l$ the number of data points with zero labels that ends in the left leaf
- $c_1^l$ the number of data points with one label that ends in the left leaf
- $c_0^r$ the number of data points with zero labels that ends in  the right leaf
- $c_1^r$ the number of data points with one label that ends in the right leaf

If we know these values for a given root node (split) then we can compute the cost of that split simply as $\frac{1}{n} \cdot (\min(c_0^l, c_1^l) + \min(c_0^r, c_0^r))$
since the min of the values for a leaf is the number of mispredictions in the data from that leaf. Summing over the leaves gives the total number of mispredictions with the current split.
The normalizing $1/n$ is irrelevant for determining the best split!

Now we need to show how to maintain these as the algorithm progresses. Initially, all data is in one of the leaves and the other one is empty.
When we scan the different root nodes/splits in sorted order, each step only changes these values by at most one depending on the data point we move from one leaf to the other and we can simply update the counts.

### End Solution
